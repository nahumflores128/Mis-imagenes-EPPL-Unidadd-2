Este programa estГЎ diseГұado para estudiantes de preparatoria que inician en el mundo de la programaciГіn con Python. El objetivo es crear una aplicaciГіn interactiva que permita calcular las propiedades fundamentales de un cГӯrculo y una esfera a partir del radio.

El programa calcula tres valores geomГ©tricos importantes: el ГЎrea del cГӯrculo (superficie), el perГӯmetro (longitud de la circunferencia) y el volumen de una esfera (figura tridimensional). Cada cГЎlculo se realiza en funciones independientes que solicitan el radio al usuario cuando son llamadas.

![elcirculo](https://github.com/nahumflores128/Mis-imagenes-EPPL-Unidadd-2/blob/main/elcirculo.PNG?raw=true)

El programa se ha diseГұado siguiendo un enfoque estrictamente procedural, alineado con el nivel de conocimiento esperado para un estudiante de preparatoria que inicia en Python. Se han definido cuatro funciones principales con responsabilidades claramente delimitadas:

*   **`areacirculo()`**: Encargada exclusivamente de solicitar el radio, aplicar la fГіrmula `A = ПҖ * rВІ` y mostrar el resultado formateado.
*   **`perimetrocirculo()`**: Gestiona la solicitud del radio, aplica `P = 2 * ПҖ * r` y presenta la longitud de la circunferencia.
*   **`volumencirculo()`**: Calcula el volumen de la esfera mediante `V = (4/3) * ПҖ * rВі`, extendiendo el concepto a tres dimensiones.
*   **`menu()`**: ActГәa como el nГәcleo de control del programa. Presenta una interfaz visual enriquecida, captura la elecciГіn del usuario, dirige el flujo hacia la funciГіn correspondiente y mantiene el ciclo de ejecuciГіn activo hasta que se solicite la salida.

Cada funciГіn de cГЎlculo solicita el radio internamente, lo que simplifica la firma de las funciones y evita la necesidad de gestionar el paso de argumentos en esta etapa de aprendizaje. La funciГіn `menu()` utiliza un bucle `while` controlado por una variable booleana (`continuar`), garantizando que el programa no termine abruptamente y permita una experiencia de usuario fluida y reutilizable.

```python
# ============================================================================
# рҹҺ“ PROGRAMA DE CГҒLCULOS GEOMГүTRICOS - NIVEL PREPARATORIA
# ============================================================================
# Autor: Estudiante de Preparatoria
# DescripciГіn: Calcula ГЎrea, perГӯmetro de un cГӯrculo y volumen de una esfera
# Fecha: 2026
# ============================================================================

import math  # Importamos la librerГӯa math para usar el valor de pi

# ============================================================================
# FUNCIГ“N: areacirculo()
# DescripciГіn: Calcula el ГЎrea de un cГӯrculo
# FГіrmula: A = ПҖ Г— rВІ
# ============================================================================
def areacirculo():
    print("\n" + "="*60)
    print("рҹ”ө CГҒLCULO DEL ГҒREA DEL CГҚRCULO")
    print("="*60)
    print("FГіrmula: A = ПҖ Г— rВІ")
    print("-"*60)

    radio = float(input("вңҸпёҸ  Ingresa el valor del radio (r): "))
    area = math.pi * (radio ** 2)

    print("\nрҹ“Ҡ RESULTADOS:")
    print("-"*60)
    print(f"   Radio ingresado:    {radio}")
    print(f"   Valor de ПҖ (pi):    {math.pi}")
    print(f"   рҹҺҜ ГҒrea del cГӯrculo: {area:.4f} unidadesВІ")
    print("="*60)
    input("\nрҹ’Ў Presiona ENTER para continuar...")

# ============================================================================
# FUNCIГ“N: perimetrocirculo()
# DescripciГіn: Calcula el perГӯmetro (circunferencia) de un cГӯrculo
# FГіrmula: P = 2 Г— ПҖ Г— r
# ============================================================================
def perimetrocirculo():
    print("\n" + "="*60)
    print("рҹ”ҙ CГҒLCULO DEL PERГҚMETRO DEL CГҚRCULO")
    print("="*60)
    print("FГіrmula: P = 2 Г— ПҖ Г— r")
    print("-"*60)

    radio = float(input("вңҸпёҸ  Ingresa el valor del radio (r): "))
    perimetro = 2 * math.pi * radio

    print("\nрҹ“Ҡ RESULTADOS:")
    print("-"*60)
    print(f"   Radio ingresado:        {radio}")
    print(f"   Valor de ПҖ (pi):        {math.pi}")
    print(f"   рҹҺҜ PerГӯmetro del cГӯrculo: {perimetro:.4f} unidades")
    print("="*60)
    input("\nрҹ’Ў Presiona ENTER para continuar...")

# ============================================================================
# FUNCIГ“N: volumencirculo()
# DescripciГіn: Calcula el volumen de una esfera
# FГіrmula: V = (4/3) Г— ПҖ Г— rВі
# ============================================================================
def volumencirculo():
    print("\n" + "="*60)
    print("рҹҹў CГҒLCULO DEL VOLUMEN DE LA ESFERA")
    print("="*60)
    print("FГіrmula: V = (4/3) Г— ПҖ Г— rВі")
    print("-"*60)

    radio = float(input("вңҸпёҸ  Ingresa el valor del radio (r): "))
    volumen = (4/3) * math.pi * (radio ** 3)

    print("\nрҹ“Ҡ RESULTADOS:")
    print("-"*60)
    print(f"   Radio ingresado:      {radio}")
    print(f"   Valor de ПҖ (pi):      {math.pi}")
    print(f"   рҹҺҜ Volumen de la esfera: {volumen:.4f} unidadesВі")
    print("="*60)
    input("\nрҹ’Ў Presiona ENTER para continuar...")

# ============================================================================
# FUNCIГ“N: menu()
# DescripciГіn: Muestra el menГә principal y controla el flujo del programa
# ============================================================================
def menu():
    continuar = True

    while continuar:
        print("\n" + "\n" * 2)

        print("в•”" + "в•җ"*58 + "в•—")
        print("в•‘" + " "*15 + "рҹҺ“ GEOMETRГҚA INTERACTIVA рҹҺ“" + " "*15 + "в•‘")
        print("в•ҡ" + "в•җ"*58 + "в•қ")
        print()
        print("в”Ң" + "в”Җ"*58 + "в”җ")
        print("в”Ӯ" + " "*18 + "MENГҡ PRINCIPAL" + " "*18 + "в”Ӯ")
        print("в”ң" + "в”Җ"*58 + "в”Ө")
        print("в”Ӯ" + " "*58 + "в”Ӯ")
        print("в”Ӯ  рҹ”ө 1. Calcular ГҒREA del cГӯrculo                  в”Ӯ")
        print("в”Ӯ  рҹ”ҙ 2. Calcular PERГҚMETRO del cГӯrculo             в”Ӯ")
        print("в”Ӯ  рҹҹў 3. Calcular VOLUMEN de la esfera              в”Ӯ")
        print("в”Ӯ  рҹҡӘ 4. SALIR del programa                         в”Ӯ")
        print("в”Ӯ" + " "*58 + "в”Ӯ")
        print("в””" + "в”Җ"*58 + "в”ҳ")
        print()

        opcion = input("рҹ‘ү Selecciona una opciГіn (1-4): ")
        print()

        if opcion == "1":
            areacirculo()
        elif opcion == "2":
            perimetrocirculo()
        elif opcion == "3":
            volumencirculo()
        elif opcion == "4":
            print("\n" + "вңЁ"*30)
            print("рҹ‘Ӣ ВЎGracias por usar el programa!")
            print("рҹҺ“ ВЎHasta pronto, estudiante!")
            print("вңЁ"*30 + "\n")
            continuar = False
        else:
            print("вҡ пёҸ  OPCIГ“N NO VГҒLIDA")
            print("рҹ’Ў Por favor, selecciona una opciГіn del 1 al 4")
            print("-"*60)
            input("рҹ’Ў Presiona ENTER para intentar de nuevo...")

# ============================================================================
# PROGRAMA PRINCIPAL
# ============================================================================
if __name__ == "__main__":
    menu()
```
